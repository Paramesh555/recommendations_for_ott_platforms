```python

!wget -O moviedataset.zip https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%205/data/moviedataset.zip
print('unziping ...')

!unzip -o -j moviedataset.zip 
###################
import pandas as pd
#Math functions, we'll only need the sqrt function so let's import only that
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
###################
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('ratings.csv')

movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))',expand=False)

movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand=False)

movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '')

movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
######################
# genre split
movies_df['genres'] = movies_df.genres.str.split('|') 
######################
# to get genre in 0 and 1 
moviesWithGenres_df = movies_df.copy()
for index, row in movies_df.iterrows():
    for genre in row['genres']:
        moviesWithGenres_df.at[index, genre] = int(1)
       
moviesWithGenres_df = moviesWithGenres_df.fillna(0)

userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ] 
inputMovies = pd.DataFrame(userInput)

inputMovies = pd.merge(moviesWithGenres_df, inputMovies)
inputMovies.drop('title',axis=1,inplace=True)
df_rat=inputMovies['rating']
inputMovies.drop('(no genres listed)',axis=1,inplace=True)
inputMovies.drop('movieId',axis=1,inplace=True)
inputMovies.drop('genres',axis=1,inplace=True)
inputMovies.drop('year',axis=1,inplace=True)
inputMovies.drop('rating',axis=1,inplace=True)

arr=inputMovies.to_numpy()
df_rat1=df_rat.T
arr1=df_rat1.to_numpy()

a=len(inputMovies.columns)
b=len(arr1)

z=[]

for i in range(0,a):
     sum=0
     for j in range(0,b):
          sum+=arr1[j]*arr[j][i]

     z.append(sum)
# print(z)
s=0
for i in range(0,len(z)):
        s+=z[i]
# print(s)
for i in range(0,len(z)):
        z[i]=round((z[i]/s),5)

moviesWithGenres_df.drop(0,axis=0,inplace=True)
moviesWithGenres_df.drop(1,axis=0,inplace=True)
moviesWithGenres_df.drop(295,axis=0,inplace=True)
moviesWithGenres_df.drop(1273,axis=0,inplace=True)
moviesWithGenres_df.drop(1967,axis=0,inplace=True)
moviesWithGenres_df.drop('(no genres listed)',axis=1,inplace=True)
moviesWithGenres_df.drop('movieId',axis=1,inplace=True)
moviesWithGenres_df.drop('genres',axis=1,inplace=True)
moviesWithGenres_df.drop('year',axis=1,inplace=True)
moviesWithGenres_df=moviesWithGenres_df.reset_index()
moviesWithGenres_df.drop('index',axis=1,inplace=True)
new_df=moviesWithGenres_df


newdf=new_df.copy()


moviesWithGenres_df.drop('title',axis=1,inplace=True)

moviesWithGenres_df=moviesWithGenres_df.T
new=moviesWithGenres_df.to_numpy()
# print(new.shape)

f=len(moviesWithGenres_df.columns)
c=len(z)


q=[]

for i in range(0,f):
     sum1=0
     for j in range(0,c):
          sum1+=z[j]*new[j][i]

     q.append(sum1)

for i in range(0,len(q)):
        q[i]=round((q[i]*5),5)

e=[]
for i in range(0,len(q)):
  e.append(q[i])

q.sort(reverse=True)
h=[]
for i in range(0,8):
        index=e.index(q[i])
        h.append(index)
        
# print(h)   #indexes of maximums

pes=[]
for i in h:
  if i not in pes:
        pes.append(i)

print("\n")
print("########################################################################################################################################################################################################")
print("YOU MIGHT ALSO LIKE")
for i in range(0,len(pes)):
    print(i+1,".",newdf.loc[pes[i],'title'])
print("########################################################################################################################################################################################################")


```
